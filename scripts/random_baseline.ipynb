```python
from run_generation_processing import get_average_prediction_similarity, get_accuracy_level
import numpy as np
import pandas as pd
import torch
import os

os.chdir('../')
```

```python
def get_average_accuracy(generated_ECs, true_EC):
    accuracies = []
    for ec in generated_ECs:
        accuracies.append(get_accuracy_level(ec, true_EC))
    return np.mean(accuracies)
```

```python
ec2drfp = torch.load('data/ec2drfp.pt')

with open('data/ECs_generation/train_rare_ecs.txt', 'r') as f:
    train_rare_ecs = f.read().splitlines()
with open('data/ECs_generation/test_ecs.txt', 'r') as f:
    test_ecs = f.read().splitlines()

ecs = train_rare_ecs + test_ecs

summary_df = pd.DataFrame({'ec': ecs})
options = list(ec2drfp.keys())

#sample 200 random ECs from sample
summary_df['generated_ECs'] = summary_df.apply(lambda x: np.random.choice(options, 200), axis=1)

summary_df['Similarity to Target EC'] = summary_df.apply(lambda x: get_average_prediction_similarity(x['ec'], x['generated_ECs'], ec2drfp), axis=1)
summary_df['average_accuracy_level'] = summary_df.apply(lambda x: get_average_accuracy(x['generated_ECs'], x['ec']), axis=1)
#calculate how many of the generated ECs are correct
summary_df['Correct EC'] = summary_df.apply(lambda x: sum([1 for ec in x['generated_ECs'] if ec == x['ec']])/200, axis=1)
```

```python
summary_df['split'] = summary_df['ec'].apply(lambda x: 'Train Rare ECs' if x in train_rare_ecs else 'Heldout ECs')
summary_df['Model'] = 'Random'
summary_df
```

```python
summary_df.to_csv('results/random_baseline.csv', index=False)
```
